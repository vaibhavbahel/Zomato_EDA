```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from termcolor import cprint
```

```python
data = pd.read_csv('zomato.csv')
cprint("TAIL of zomatao dataframe\n\n","blue",attrs=["bold"])
data.tail()
```

```python
cprint("Information about data frame \n\n","red",attrs=["bold"])
data.info()
```

```python
data = data.drop(['url','dish_liked','phone'],axis=1)
cprint("\nDropped url,dish_liked and phone columns\n","blue",attrs=["bold"])
```

```python
print("Number of duplicate values: {}".format(data.duplicated().sum()))
data.drop_duplicates(inplace=True)
print("Duplicate values dropped")
print("Number of duplicate values after drop: {}".format(data.duplicated().sum()))
```

```python
print("Number of null values:\n\n{}".format(data.isnull().sum()))
data.dropna(how='any',inplace=True)
print("Number of null values after drop:\n\n{}".format(data.isnull().sum()))
```

```python
data.info()
```

```python
print('columns before rename:\n\n{}'.format(data.columns))
```

```python
data = data.rename(columns={'approx_cost(for two people)':'cost', 'listed_in(type)':'type','listed_in(city)':'city'})

print('columns after rename:\n\n{}'.format(data.columns))
```

```python
data['cost'] = data['cost'].astype('str')

data['cost'] = data['cost'].apply(lambda x:x.replace(',','.'))

data['cost'] = data['cost'].astype(float)

cprint('Info after transformation:\n\n','green',attrs=['bold'])

data.info()
```

```python
cprint('Unique values of ratings\n','green',attrs=['bold'])

data['rate'].unique() 
```

```python
data = data.loc[data.rate !='NEW']
data = data.loc[data.rate != '-'].reset_index(drop=True)
remove_slash = lambda x:x.replace('/5','') if type(x) == np.str else x
data.rate = data.rate.apply(remove_slash).str.strip().astype('float')
cprint('Ratings after transforms:\n','green',attrs=['bold'])
print(data['rate'].head())
```

```python
data.name = data.name.apply(lambda x:x.title())
data.online_order.replace(('Yes','No'),(True,False),inplace=True)
data.book_table.replace(('Yes','No'),(True,False),inplace=True)
cprint('Unique values of Cost:\n','green',attrs=['bold'])
print(data.cost.unique())
```

```python
def Encode(data):
    for column in data.columns[~data.columns.isin(['rate','cost','votes'])]:
        data[column] = data[column].factorize()[0]
    return (data)

data_en = Encode(data.copy())
```

```python
cprint('DF after encoding:\n','green',attrs=['bold'])
data_en
```

```python
corr = data_en.corr(method='kendall')
plt.figure(figsize=(15,10))
sns.heatmap(corr,annot=True)
cprint('Columns and Correlation Matrix\n','red',attrs=['bold'])
print(data_en.columns)
```

```python
cprint('Plot for restraunt table booking vs rating\n','cyan',attrs=['bold'])
plt.rcParams['figure.figsize'] = (13,10)
Y = pd.crosstab(data['rate'],data['book_table'])
Y.div(Y.sum(1).astype(float),axis=0).plot(kind='bar',stacked=True,color=['red','yellow'])
plt.title('Table booking vs rate',fontweight=30,fontsize=20)
plt.legend(loc='upper right')
plt.show()
```
